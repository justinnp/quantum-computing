```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
```

## Q1.A

```python
qc = QuantumCircuit(3)
# H gate at q0
qc.h(0)
# CNOT with q0 as control and q1 as target
qc.cx(0,1)
qc.cx(1,2)
qc.draw("mpl")
```
